# A visualisation of the SIR Model


```python
# Import functions from sir_model script
from sir_model import sir_model, simulate_sir

# Example parameters (you can change these)

I0 = 0.01        # Initial proportion of infected individuals
R0 = 0           # Initial proportion of recovered individuals
S0 = 1 - I0 - R0 # Initial proportion of susceptible individuals
beta = 0.1       # Infection rate
gamma = 0.07     # Recovery rate
days = 160       # Number of days to simulate


# Plot the results of the SIR model simulation

results = simulate_sir(S0, I0, R0, beta, gamma, days)

# Explore parameter changes
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot([day for day, (S, I, R) in enumerate(results)], [S for day, (S, I, R) in enumerate(results)], label='Susceptible', color='blue')
plt.plot([day for day, (S, I, R) in enumerate(results)], [I for day, (S, I, R) in enumerate(results)], label='Infected', color='red')
plt.plot([day for day, (S, I, R) in enumerate(results)], [R for day, (S, I, R) in enumerate(results)], label='Recovered', color='green')
plt.title('SIR Model Simulation')
plt.xlabel('Days')
plt.ylabel('Proportion of Population')
plt.legend()
plt.grid()
```
