# A visualisation of the SIR Model

Here I have created a basic representantion of the SIR model. The SIR model is a common model used to show how a population my react to differnet diseases based on the severity of them. The model takes a population of size N and splits them into 3 catagories:

S(t) - Those who are still __susceptible__ to the disease.

I(t) - Those who are __infectious__.

R(t) - Those who have __recovered__ and are no longer infectious.

The term R(t) can also be thought of as removed, accounting for those who have died from the disease. These functions of time are governed by the 3 ordinary differential equations:

$\frac{dS}{dt} = -\beta \frac{SI}{N}$

$\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I$

$\frac{dR}{dt} = \gamma I$


```python
# Import functions from sir_model script
from sir_model import sir_model, simulate_sir

# Example parameters (you can change these)

I0 = 0.01        # Initial proportion of infected individuals
R0 = 0           # Initial proportion of recovered individuals
S0 = 1 - I0 - R0 # Initial proportion of susceptible individuals
beta = 0.2       # Infection rate
gamma = 0.15      # Recovery rate
days = 160       # Number of days to simulate


# Plot the results of the SIR model simulation

results = simulate_sir(S0, I0, R0, beta, gamma, days)

# Explore parameter changes
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot([day for day, (S, I, R) in enumerate(results)], [S for day, (S, I, R) in enumerate(results)], label='Susceptible', color='blue')
plt.plot([day for day, (S, I, R) in enumerate(results)], [I for day, (S, I, R) in enumerate(results)], label='Infected', color='red')
plt.plot([day for day, (S, I, R) in enumerate(results)], [R for day, (S, I, R) in enumerate(results)], label='Removed', color='green')
plt.title('SIR Model Simulation')
plt.xlabel('Days')
plt.ylabel('Proportion of Population')
plt.legend()
plt.grid()
```

Here we have made a plot covering 160 days an epidemic in order to get a feel for how each of these variables interact with one another over time as well as how changing the parameters $\beta$ and $\gamma$ affect the ultimate number of cases and the speed with which the epidemic spreads. 
Another number not yet mentioned is the basic reproduction number, $R_0$ = $\frac{\beta}{\gamma}$. This basic reproductive rate, while simple, can tell us a lot about how infectious the epidemic at hand can be.
